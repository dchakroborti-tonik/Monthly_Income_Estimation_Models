```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.cloud import bigquery
import os
# Connection to Bigquery
client = bigquery.Client(project='prj-prod-dataplatform')
```

```python
sq = """
SELECT
  digitalLoanAccountId,
  loan_salary_scaled_income,
  loan_monthly_income
FROM
  worktable_data_analysis.Demographic_Beta1_Jan2023_dec2024
  ;
"""

dfloan = client.query(sq).to_dataframe(progress_bar_type = 'tqdm')
print(f"The shape of dfloan is:\t {dfloan.shape}")
```

```python
resultdf = pd.read_csv(r"/home/jupyter/Models/Beta1_IncomeEstimation_RegressionModel/Models/Beta1V4.0/Step1/result_df_Top10SHAPFeatures_Beta1V4_Step1.csv")
print(f"The shape of result df is:\t {resultdf.shape}")
```

```python
df = dfloan.merge(resultdf, on = 'digitalLoanAccountId', how = 'inner')
print(f"The shape of df is:\t {df.shape}")
```

```python
df.head()
```

```python
df = df[['digitalLoanAccountId', 'loan_salary_scaled_income', 'loan_monthly_income', 'actual_value', 'predicted_value']].copy()
```

```python
# First convert to float, then to int
df['loan_salary_scaled_income'] = df['loan_salary_scaled_income'].astype(float).astype(np.int64)
df['loan_monthly_income'] = df['loan_monthly_income'].astype(float).astype(np.int64)
df['actual_value'] = df['actual_value'].astype(float).astype(np.int64)
df['predicted_value'] = df['predicted_value'].astype(float).astype(np.int64)
df.head(10)
```

```python
df.isnull().sum()
```

```python
import pandas as pd
import numpy as np

# Create bins from 0 to 180000 with 20000 intervals
bins = np.arange(0, 252000, 10000)

# Create labels with alphabetical prefixes
# Using chr(97 + i) to generate letters a, b, c, etc.
labels = [f'{chr(97 + i)}_{bins[i]}-{bins[i+1]}' for i in range(len(bins)-1)]

# Create binned columns for each metric
df['scaled_income_bins'] = pd.cut(df['loan_salary_scaled_income'], 
                                bins=bins, 
                                labels=labels, 
                                include_lowest=True)

df['monthly_income_bins'] = pd.cut(df['loan_monthly_income'], 
                                 bins=bins, 
                                 labels=labels, 
                                 include_lowest=True)

df['predicted_value_bins'] = pd.cut(df['predicted_value'], 
                                  bins=bins, 
                                  labels=labels, 
                                  include_lowest=True)

# To see the distribution in each bin
print("\nScaled Income Distribution:")
print(df['scaled_income_bins'].value_counts().sort_index())

print("\nMonthly Income Distribution:")
print(df['monthly_income_bins'].value_counts().sort_index())

print("\nPredicted Value Distribution:")
print(df['predicted_value_bins'].value_counts().sort_index())
```

```python
d1 = df.groupby(['monthly_income_bins'], observed=True)['digitalLoanAccountId'].nunique().reset_index()
d1
```

```python
d2 = df.groupby(['scaled_income_bins'], observed=True)['digitalLoanAccountId'].nunique().reset_index()
d2
```

```python
d3 = df.groupby(['predicted_value_bins'], observed=True)['digitalLoanAccountId'].nunique().reset_index()
d3
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the figure and axis
plt.figure(figsize=(12, 6))

# Plot the lines with different colors and markers
plt.plot(d1['monthly_income_bins'], d1['digitalLoanAccountId'], 'b-o', label='Monthly Income', linewidth=2)
plt.plot(d2['scaled_income_bins'], d2['digitalLoanAccountId'], 'g-s', label='Scaled Income', linewidth=2)
plt.plot(d3['predicted_value_bins'], d3['digitalLoanAccountId'], 'r-^', label='Predicted Value', linewidth=2)

# Customize the plot
plt.title('Comparison of Count of Loan Accounts across Income Bins', fontsize=14, pad=20)
plt.xlabel('Income Bins', fontsize=12)
plt.ylabel('Number of Digital Loan Accounts', fontsize=12)
plt.legend(fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Format x-axis labels to remove prefix
x_labels = [bin_label[2:].replace('-', ' - ') for bin_label in d1['monthly_income_bins']]

# Set the tick positions and labels
ax = plt.gca()
ax.set_xticks(range(len(x_labels)))  # Set the tick positions
ax.set_xticklabels(x_labels)  # Set the tick labels

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# First, let's make sure we're using the same bins across all dataframes
common_bins = d1['monthly_income_bins'].unique()

# Create a new figure
plt.figure(figsize=(15, 6))
bar_width = 0.25

# Create positions for bars
x = np.arange(len(common_bins))

# Get data for each category, ensuring they align with common_bins
d1_values = d1.set_index('monthly_income_bins')['digitalLoanAccountId'].reindex(common_bins).fillna(0)
d2_values = d2.set_index('scaled_income_bins')['digitalLoanAccountId'].reindex(common_bins).fillna(0)
d3_values = d3.set_index('predicted_value_bins')['digitalLoanAccountId'].reindex(common_bins).fillna(0)

# Create the bars
plt.bar(x - bar_width, d1_values, bar_width, label='Monthly Income', alpha=0.8)
plt.bar(x, d2_values, bar_width, label='Scaled Income', alpha=0.8)
plt.bar(x + bar_width, d3_values, bar_width, label='Predicted Value', alpha=0.8)

# Customize x-axis labels
x_labels = [bin_label[2:].replace('-', ' - ') for bin_label in common_bins]
plt.xticks(x, x_labels, rotation=45, ha='right')

# Add title and labels
plt.title('Comparison of Digital Loan Accounts across Income Bins')
plt.xlabel('Income Bins')
plt.ylabel('Number of Digital Loan Accounts')

# Add legend
plt.legend()

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
